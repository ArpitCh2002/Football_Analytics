```python
# Importing the packages

import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# Reading the data

df = pd.read_csv('D:\\Power BI\\Sports\\Football Pitch\\Dataset\\xgchart.csv')
```

```python
# Displaying the dataframe

df
```

```python
# Creating lists for xG chart creation

h_xG = [0]
a_xG = [0]
h_min = [0]
a_min = [0]

# Finding home and away teams

hteam = df['team'].iloc[0]
ateam = df['team'].iloc[-1]

for x in range(len(df['xG'])):
    if df['team'][x] == ateam:
        a_xG.append(df['xG'][x])
        a_min.append(df['minute'][x])
    if df['team'][x] == hteam:
        h_xG.append(df['xG'][x])
        h_min.append(df['minute'][x])
```

```python
a_xG
```

```python
# Creating a cumulative value function

def nums_cumulative_sum(nums_list):
    return [sum(nums_list[:i+1]) for i in range(len(nums_list))]

a_cumulative = nums_cumulative_sum(a_xG)
h_cumulative = nums_cumulative_sum(h_xG)
```

```python
a_total = round(a_cumulative[-1],2)
h_total = round(h_cumulative[-1],2)
```

```python
# Plotting the xG flowchart for data

fig, ax = plt.subplots(figsize=(10,5))
fig.set_facecolor('#3d4849')
ax.patch.set_facecolor('#3d4849')

plt.xticks([0,15,30,45,60,75,90])
plt.xlabel('Match Minutes', color='#FFFFFF')
plt.ylabel('xG', color='#FFFFFF')

plt.gca().spines['top'].set_color('white')
plt.gca().spines['bottom'].set_color('white')
plt.gca().spines['left'].set_color('white')
plt.gca().spines['right'].set_color('white')

plt.gca().tick_params(axis='x', colors='white')
plt.gca().tick_params(axis='y', colors='white')

ax.step(x=a_min, y=a_cumulative, where='post')
ax.step(x=h_min, y=h_cumulative, where='post')
```
