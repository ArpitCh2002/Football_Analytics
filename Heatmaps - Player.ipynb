```python
import pandas as pd
import matplotlib.pyplot as plt
import mplsoccer.pitch as Pitch
import seaborn as sns
```

```python
# Read the dataset

df = pd.read_csv('D:\Power BI\Sports\Football Pitch\Dataset\messibetis.csv')
```

```python
# Print the dataset

df
```

```python
# Aligning pass coordinates with Pitch

df['x'] = df['x']*1.2
df['y'] = df['y']*0.8

df['endX'] = df['endX']*1.2
df['endY'] = df['endY']*0.8
```

```python
#Print the adjusted dataframe

df
```

```python
import mplsoccer
print(mplsoccer.__version__)
```

```python
# Creating the Pitch
from mplsoccer.pitch import Pitch
import matplotlib.pyplot as plt

pitch = Pitch(pitch_type='statsbomb', pitch_color='#22312b', line_color='#c7d5cc')
fig, ax = pitch.draw(figsize=(13.5, 8), constrained_layout=True, tight_layout=False)
fig.set_facecolor('#22312b')

# Inverting Pitch to align with Passes
plt.gca().invert_yaxis()

# Creating Pass Map
for x in range(len(df['x'])):
    if df['outcome'][x] == 'Successful':
        plt.plot((df['x'][x], df['endX'][x]) , (df['y'][x], df['endY'][x]) , color='green')
        plt.scatter(df['x'][x] , df['y'][x] , color='green')
    if df['outcome'][x] == 'Unsuccessful':
        plt.plot((df['x'][x], df['endX'][x]) , (df['y'][x], df['endY'][x]) , color='red')
        plt.scatter(df['x'][x] , df['y'][x] , color='red')
    
plt.title('Messi Pass Map vs Real Betis', color='white', size=24)
```
