# Player Dashboard Project

### In this project, we will create a dashboard which can be used to show data for club, country or players usign Python. Along with Python, we will use GridSpec to make dashboard interactive.

```python
# Importing the required packages and modules

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
# printing random numbers

x = np.random.randint(1,101,500)
```

```python
# Importing the required dataset

df = pd.read_csv('D:\Power BI\Sports\Football Pitch\Dataset\europaFinal.csv')

# Printing the initial values of the dataframe
df.head()
```

```python
# Fetching specific data from df dataframe by creating functions

def plot_heatmaps(heat_df, teamId, p_type, ax):
    heat_df = heat_df.copy()
    heat_df = heat_df.loc[(heat_df['teamId'] == teamId) & (heat_df['type'] == p_type)]
    
    kde = sns.kdeplot(x=heat_df.x, y=heat_df.y, data=heat_df, fill=True, cmap='magma', ax=ax)
```

```python
# Creating the axes grid using Gridspec

fig = plt.figure(figsize=(10,10), constrained_layout=True)

gs = fig.add_gridspec(nrows=3, ncols=4)

ax1 = fig.add_subplot(gs[0,0])
ax1.set_title('[0,0]')

ax2 = fig.add_subplot(gs[0,1])
ax2.set_title('[0,1]')

ax3 = fig.add_subplot(gs[0,2])
ax3.set_title('[0,2]')

ax4 = fig.add_subplot(gs[0,3])
ax4.set_title('[0,3]')

ax5 = fig.add_subplot(gs[1,0])
ax5.set_title('[1,0]')

ax6 = fig.add_subplot(gs[1,1])
ax6.set_title('[1,1]')

ax7 = fig.add_subplot(gs[1,2])
ax7.set_title('[1,2]')

ax8 = fig.add_subplot(gs[1,3])
ax8.set_title('[1,3]')

ax9 = fig.add_subplot(gs[2,0])
ax9.set_title('[2,0]')

ax10 = fig.add_subplot(gs[2,1])
ax10.set_title('[2,1]')

ax11 = fig.add_subplot(gs[2,2])
ax11.set_title('[2,2]')

ax12 = fig.add_subplot(gs[2,3])
ax12.set_title('[2,3]')
```

```python
# Customizing the created axes grid using Gridspec

fig = plt.figure(figsize=(10,10), constrained_layout=True)

gs = fig.add_gridspec(nrows=3, ncols=4)

ax1 = fig.add_subplot(gs[0,:2])
ax1.set_title('[0,:2]')

ax3 = fig.add_subplot(gs[0,2])
ax3.set_title('[0,2]')

ax4 = fig.add_subplot(gs[0,3])
ax4.set_title('[0,3]')

ax5 = fig.add_subplot(gs[1:,0])
ax5.set_title('[1:,0]')

ax6 = fig.add_subplot(gs[1,1])
ax6.set_title('[1,1]')

ax7 = fig.add_subplot(gs[1,2])
ax7.set_title('[1,2]')

ax8 = fig.add_subplot(gs[1,3])
ax8.set_title('[1,3]')

ax10 = fig.add_subplot(gs[2,1])
ax10.set_title('[2,1]')

ax11 = fig.add_subplot(gs[2,2])
ax11.set_title('[2,2]')

ax12 = fig.add_subplot(gs[2,3])
ax12.set_title('[2,3]')
```

```python
# Plotting data into the Customized axes grid using Gridspec

fig = plt.figure(figsize=(10,10), constrained_layout=True)

gs = fig.add_gridspec(nrows=3, ncols=4)

ax1 = fig.add_subplot(gs[0,:2])
ax1.set_title('[0,:2]')
boxplot = sns.boxplot(x=x, ax=ax1)

ax3 = fig.add_subplot(gs[0,2])
ax3.set_title('[0,2]')
plot_heatmaps(heat_df=df, teamId='Villarreal', p_type='Pass', ax=ax3)

ax4 = fig.add_subplot(gs[0,3])
ax4.set_title('[0,3]')
plot_heatmaps(heat_df=df, teamId='Manchester United', p_type='Pass', ax=ax4)

ax5 = fig.add_subplot(gs[1:,0])
ax5.set_title('[1:,0]')

ax6 = fig.add_subplot(gs[1,1])
ax6.set_title('[1,1]')

ax7 = fig.add_subplot(gs[1,2])
ax7.set_title('[1,2]')

ax8 = fig.add_subplot(gs[1,3])
ax8.set_title('[1,3]')

ax10 = fig.add_subplot(gs[2,1])
ax10.set_title('[2,1]')

ax11 = fig.add_subplot(gs[2,2])
ax11.set_title('[2,2]')

ax12 = fig.add_subplot(gs[2,3])
ax12.set_title('[2,3]')
ax12.scatter(5,2)
```
